#### . There are no non_values
#### . town is a nominal categorical variable

```python
### Apply get_dummies() using pandas to transform the nominal categorical variable(Town).

data_1 = pd.get_dummies(data.town)
data_1
```

```python
data_2 = pd.concat([data, data_1], axis='columns')
data_2
```

```python
data_2.drop('town', axis='columns', inplace=True)
data_2
```

```python
### Assume that Town is a ordinal categorical variable.
### Now apply map() function using pandas.

df = data.copy()
df['town'] = df['town'].map({'Bangalore':0, 'Chennai':1, 'Hyderabad':2})
df
```
